```python
import matplotlib.pyplot as plt

# Step 1: Original square points (corners)
square = [(1, 1), (1, 4), (4, 4), (4, 1)]

# Step 2: Translation values (how much to move)
move_right = 4
move_up = 2

# Step 3: Translate the square (add move_right to x, move_up to y)
new_square = []
for point in square:
    x, y = point
    new_x = x + move_right
    new_y = y + move_up
    new_square.append((new_x, new_y))

# Step 4: To draw a closed square, add the first point again at the end
square.append(square[0])
new_square.append(new_square[0])

# Step 5: Separate x and y for plotting
x1, y1 = zip(*square)      # Original square
x2, y2 = zip(*new_square)  # Translated square

# Step 6: Plot both squares
plt.plot(x1, y1, 'go-', label='Original Square')   # green with circles
plt.plot(x2, y2, 'ro-', label='Translated Square') # magenta with circles

# Extra: Add grid and axis
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Translation of a Square")
plt.legend()
plt.axis('equal')  # Keep square shape
plt.show()
```
